```python
import requests
import schedule
import time
from plyer import notification
import pytz
from datetime import datetime

# https://www.youtube.com/watch?v=FRIyIlsHUXM&t=15s


def get_weather():
    api_key = "3008d41862772406dd614cc6a679ba67"
    city_name = "New Rochelle"
    base_url = "https://api.openweathermap.org/data/2.5/weather?"
    
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name

    # Get weather using lat and lon
    # lat = "40.90"
    # lon = "-73.77"
    # complete_url = base_url + "lat=" + lat + "&lon=" + lon + "&appid=" + api_key

    response = requests.get(complete_url)
    data = response.json()
    
    if data['cod'] != "404":
        main = data["main"]
        current_temperature = main["temp"]
        current_temp_celsius = round(current_temperature - 273.15, 2)
        current_temp_fah = round((current_temperature - 273.15) *9/5 + 32, 2)

        return current_temp_celsius, current_temp_fah, city_name


# def get_forecast():
#     # api.openweathermap.org/data/2.5/forecast/daily?lat={lat}&lon={lon}&cnt={cnt}&appid={API key}
#     api_key = "3008d41862772406dd614cc6a679ba67"
#     city_name = "New Rochelle"
#     days = 5
#     base_url = "https://api.openweathermap.org/data/2.5/forecast/daily?"    
#     complete_url = base_url + "appid=" + api_key + "&q=" + city_name
#     return complete_url

def job():
    temp_c, temp_f, city= get_weather()
    notification.notify(
        title="Current Weather Update",
        message=f"Temperature is {temp_c} degrees C / {temp_f} degrees F in {city}",
        timeout=10
    )

# schedule.every().day.at("08:55").do(job)

# while True:
#     current_time = datetime.now(pytz.timezone('US/Eastern')).strftime('%H:%M')
#     if current_time == "08:55":
#         schedule.run_pending()
#     time.sleep(60)

job()
```
